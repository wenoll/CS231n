## Inline Question 1:

We've only asked you to implement ReLU, but there are a number of different activation functions that one could use in neural networks, each with its pros and cons. In particular, an issue commonly seen with activation functions is getting zero (or close to zero) gradient flow during backpropagation. Which of the following activation functions have this problem? If you consider these functions in the one dimensional case, what types of input would lead to this behaviour?
1. Sigmoid
2. ReLU
3. Leaky ReLU

$\\color{blue}{\textit Your Answer:}$

**1. Sigmoid (S型激活函数)**

**问题：** 有。

**导致该行为的输入类型：** 当输入 $x$ 变得非常大（正无穷大）或非常小（负无穷大）时。在这种情况下，Sigmoid 函数会饱和，其导数会变得非常接近于零。这意味着流经该神经元的梯度会非常小，导致其权重更新微乎其微，从而引发梯度消失问题。

**2. ReLU (修正线性单元)**

**问题：** 有（被称为“垂死ReLU”问题）。

**导致该行为的输入类型：** 当输入 $x$ 小于或等于零时。对于 $x \leq 0$，ReLU 的输出为零，其导数也为零。这意味着如果一个ReLU神经元在训练过程中始终接收到负值或零输入，它将永远不会被激活，并且其权重将不会通过反向传播进行更新，从而导致该神经元“死亡”。

**3. Leaky ReLU (渗漏修正线性单元)**

**问题：** 基本上没有。

**导致该行为的输入类型：** Leaky ReLU 的设计目的就是为了解决 ReLU 的“垂死”问题。当输入 $x$ 小于零时，Leaky ReLU 的导数是一个小的正数（例如0.01），而不是零。因此，即使对于负输入，梯度也仍然会流动，从而有效缓解了梯度消失的问题。它通常不会在任何输入类型下出现梯度接近于零的问题，除非其“泄漏”参数本身设置为零（此时就退化为普通ReLU）。

## Inline Question 2:

Now that you have trained a Neural Network classifier, you may find that your testing accuracy is much lower than the training accuracy. In what ways can we decrease this gap? Select all that apply.

1. Train on a larger dataset.
2. Add more hidden units.
3. Increase the regularization strength.
4. None of the above.

$\color{blue}{	Your Answer:}$ **1, 3**

$\color{blue}{	Your Explanation:}$

当测试准确率远低于训练准确率时，这通常表明模型出现了过拟合（Overfitting）。过拟合意味着模型在训练数据上学习得太好，包括了噪声和特定样本的细节，导致其泛化能力差，在未见过的数据（测试集）上表现不佳。

针对过拟合问题，我们可以采取以下策略来减小训练与测试准确率之间的差距：

1.  **训练更大的数据集：** 增加训练数据的多样性和数量可以帮助模型学习到更普遍、更鲁棒的模式，而不是仅仅记住训练集中的特定样本。这使得模型更难以过度拟合单一的数据点或噪声，从而提高了其泛化能力。

2.  **增加正则化强度：** 正则化是一种惩罚模型复杂度的技术。例如，L1或L2正则化通过在损失函数中添加权重大小的惩罚项，鼓励模型使用更小的权重，从而降低模型的复杂度，减少对训练数据的过度依赖。Dropout也是一种有效的正则化方法，通过随机丢弃神经元来防止网络过拟合。增加正则化强度会迫使模型学习更简单的表示，从而改善泛化性能。

**不正确的选项解释：**

*   **增加隐藏单元：** 增加隐藏单元会增加模型的容量（即模型的复杂程度），使其能够拟合更复杂的函数。如果模型已经过拟合，增加容量只会让它更严重地过拟合训练数据，导致在测试集上的表现更差，而不是减小差距。
